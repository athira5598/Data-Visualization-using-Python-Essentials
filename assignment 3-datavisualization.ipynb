###  Import Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fmri=sns.load_dataset('fmri')
fmri.head()
```

### Regression plot

```python
sns.relplot(x='timepoint',y='signal',hue='event',data=fmri,aspect=3,kind='line')
```

```python
sns.relplot(x='timepoint',y='signal',hue='region',data=fmri,aspect=2,palette='PuRd_r')
```

```python
sns.relplot(x='timepoint',y='signal',hue='region',data=fmri,aspect=2,palette='PuRd_r',kind='line')
```

```python
sns.relplot(x='timepoint',y='signal',hue='event',data=fmri,aspect=2,palette='PuRd_r',kind='line')
```

```python
sns.boxplot(x='timepoint',y='signal',hue='event',data=fmri,palette='PuRd_r',fliersize=4)
```

```python
sns.boxplot(x='timepoint',y='signal',hue='region',data=fmri,palette='PuRd_r',fliersize=3)
```

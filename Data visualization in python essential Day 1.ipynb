```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
## LINE PLOT
```

```python
plt.plot(a,b,'c--v',linewidth=2,markersize=6.5,markerfacecolor='g')
plt.xlabel('value range from 40 to 50')
plt.ylabel('value range from 50 to 60')
plt.title('Line plot')
```

```python
days=[1,2,3,4,5,6,7]
sales_1=[160,150,140,145,175,165,180]
sales_2=[70,90,160,150,140,145,175]
```

```python
plt.plot(days,sales_1,'orange',label='sales_1')
plt.plot(days,sales_2,'blue',label='sales_2')
plt.xlabel('Days',color='green')
plt.ylabel('Sales',color='green')
plt.title('Days vs Sales plot',color='green')
plt.legend(loc='lower right')
```
